```python
from dqn.dqn_classic import dqn, save, load
from utils.plot import draw
```

```python
agents_num = 9
agents_rewards_cartpole = []
av_max_qs_cartpole = []
first = True
for i in range(agents_num):
    print("agent", i+1)
    qnn, rewards, max_qs = dqn('CartPole-v0')
    agents_rewards_cartpole.append(rewards)
    av_max_qs_cartpole.append(max_qs)
    if first:
        nn_cartpole = qnn
        first = False
    print("===============================")
```

```python
save('savings/cartpole', agents_rewards_cartpole, av_max_qs_cartpole, nn_cartpole)
```

```python
agents_rewards_cartpole, av_max_qs_cartpole = load('savings/cartpole')
```
